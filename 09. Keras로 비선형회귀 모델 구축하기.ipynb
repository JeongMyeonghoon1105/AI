# 9. Keras로 비선형회귀 모델 구축하기
* 이차함수 형태의 인구수 데이터 학습
* 인구수 예측

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import random
tf.compat.v1.disable_eager_execution()
```

```python
import warnings
warnings.filterwarnings(action = 'ignore')
```

## 데이터 준비

```python
df = pd.read_csv('Tables/population.csv')

x = []
y = []

x = df['Month'].values
y = df['Population'].values
```

## 데이터 시각화

```python
sb.lmplot('Month', 'Population', data = df, fit_reg = True, scatter_kws = {"s" : 3})
```

## 모델 구성 및 학습

```python
# 레이어 생성
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='elu'),
    tf.keras.layers.Dense(128, activation='elu'),
    tf.keras.layers.Dense(256, activation='elu'),
    tf.keras.layers.Dense(512, activation='elu'),
    tf.keras.layers.Dense(1024, activation='elu'),
    tf.keras.layers.Dense(2048, activation='elu'),
    tf.keras.layers.Dense(1, activation='elu'),
])

# 컴파일(모델 학습 과정 설정)
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

# 학습
model.fit(np.array(x), np.array(y), epochs=30000, verbose=1)
```

## 학습 결과 시각화

```python
index = []
data = []

for h in range(0, 1000):
    index.append(h)
    for i in model.predict(np.array([[h]])):
        for j in i:
            data.append(j)

result_df = pd.DataFrame({'index' : index, 'data' : data})
sb.lmplot('index', 'data', data = result_df, fit_reg = True, scatter_kws = {"s" : 3})
```

## 예측 결과 출력

```python
for i in model.predict(np.array([[600]])):
    for j in i:
        print(j)
```

## 최댓값 출력

```python
yMax = 0
maxIndex = 0

for h in range(0, 500):
    for i in model.predict(np.array([[h]])):
        for j in i:
            if yMax < j:
                yMax = j
                maxIndex = h

print(maxIndex, end = ' : ')
print(yMax)
```

## 결과 출력(10개 단위)

```python
for h in range(0, 520):
    for i in model.predict(np.array([[h]])):
        for j in i:
            if h % 10 == 0:
                print(h, end = '\t=>\t')
                print(j)
```
